## Question 2.2 -- Quantifying fairness in a lending scenario
rubric={reasoning:100}
    
<div class="alert alert-info">

<ol type="1">
<li>Which fairness criteria (independence, separation, or sufficiency) would you recommend that the bank uses to balance its model predictions between groups in order to prevent disparate real-world harm while being fair to qualified applicants from the different groups? Explain why you picked the one(s) you did and if there are any downsides with only focusing on that one. Also give an example of a fairness metric that satisfies this criteria (e.g. demographic parity, equal odds, equal opportunity, positive/negative predictive value).</li>

Which data will we have access to when evaluating/reviewing how the algorithm is doing? YOUR ANSWERS HERE

1. We recommend Separation as the ideal fairness criterion, given that we want the model to be fair across both groups. That is, we would expect to have protected groups (for example., gender, race, etc. ) and individuals based on their actual risk profile.  This will ensure consistency in the prediction of the model regardless of the applicant's profile and group membership.

The Independence fairness criteria follow a blind equal loan approval rate across groups. This will not take into account individuals' actual loan qualifications with a sole reference to a group membership. This criteria disregards applicant creditworthiness, which has the potential to approve risky loans that are masked within a group with lower repayment rates.  Beyond the financial instability that can arise from the potential high default rate, this will create an extra layer of fairness issues, with potentially qualified applicants being filtered out in a bid to keep similar acceptance rates across groups.

The Separation fairness criteria take the actual repayment abilities into account and ensure unbiased predictions for two individuals with the same risk profile regardless of group membership.

Examples of a fairness metric that satisfies this fairness criteria include: 

- Equal Opportunity: This metric ensures True Positive Rate (TPR) is the same across groups. That is, the approval rate for applicants with a positive risk profile (those who can repay) remains the same.
-  Equalized Odds: Balances both True Positive Rate (TPR) and False Positive Rate (FPR) between groups. That is to infer fairness across both approved loan applicants and rejected loan applicants.
  
These will minimize the impact on qualified applicants from underrepresented or disadvantaged groups while at the same time considering individual risk profiles as a prominent index in decision-making.
However, mere reliance on the separation fairness criteria can tilt sufficiency-based metrics like Positive Predictive Value (PPV). For example, a loan approval might signal a 95% repayment chance for one group but only 60% for another, making predictions harder to interpret or rely on fairly between groups. Separation also guarantees equal TPR/FPR across groups but does not ensure that TPR is high overall. A model could appear to make a fair prediction by denying loans to everyone, which would harm all qualified individuals. Thus, fairness at the individual level should also be part of the evaluation.


2. Upon implementation of the algorithm, the predicted positive class will be accessible to the bank, that is, those applicants who got the approval.  This will allow us to predict the probability of default: True Positives (TP) capturing those who got approved and repaid their loans, and False Positives (FP) capturing those who got approved but defaulted. However, the probability of default cannot be observed for the disqualifying applicants for those rejected (predicted negatives). This implies that we have unobserved default features: the False Negatives (FN) and True Negatives (TN) for the rejected loan applicants.

Fairness and performance of the model can rather be measured (especially for the unobserved False Negatives) though:

- A random approval of loans from the rejected applicants. This control group can help us observe the performance of the model depicting the frequency of rejection of qualified loan applicants.
- An appeal mechanism for second-look. Here, we can open an appeal window for rejected applicants with additional data to help understand possible bias in the model and false rejections.
- Track approval patterns over time and monitor changes in applicant demographics to identify potential fairness drift.
  
Focusing only on false positives can lead to stricter approvals, lowering defaults but unfairly rejecting more qualified applicants, so a balanced review is key. Overall, in addition to the optimal fairness criteria of minimizing the impact of unfair denials, sensitive financial decisions of this form should be complemented by continued monitoring and mechanisms to uncover unobserved errors like false negatives.
